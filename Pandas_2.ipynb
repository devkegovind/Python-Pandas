<h4><b>Convert Dictionary To DataFrame</b></h4>

```python
import pandas as pd
```

```python
d1 = {'key1' : [1, 2, 3, 4, 5, 'govind'],
      'key2': [7, 8, 9, 10, 11, 12],
      'key3': [13, 14, 15, 16, 17, 18],
      'key4': [19, 20, 21, 22, 23, 24]}
```

```python
df1 = pd.DataFrame(d1)
df1
```

```python
d2 = {'key1' : [1, 2, 3, 4, 5, "govind"],
      'key2' : [7, 8, 'devke', 10, 11, 12],
      'key3' : [13, 14, 15, 16, 17, 18],
      'key4' : [19, 20, 21, 'datascience', 'machine', 24]}
```

```python
df2 = pd.DataFrame(d2)
df2
```

<h4><b>Merge two DataFrame</b></h4>

```python
pd.merge(df1, df2)
```

<h4><b>Merge two dataframe with outer join on key4 & key1</b></h4>

```python
pd.merge(df1, df2, how = 'outer', on = ['key4', 'key1'])
```

<h4><b>To merge two dataframe with left join on key4</b></h4>

```python
pd.merge(df1, df2, how = 'left', on = 'key4')
```

<h4><b>To merge two dataframe with inner join on key4</b></h4>

```python
pd.merge(df1, df2, how = 'inner', on ='key4')
```

<h4><b>To merge two dataframe with inner join on key1</b></h4>

```python
 pd.merge(df1, df2, how = 'inner', on = 'key1')
```
