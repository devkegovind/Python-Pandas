<h3><b>Column Addition</b></h3>

Adding a New Column to an Existing DataFrame.

Adding a new column to an existing DataFrame object with column label by passing

```python
import pandas as pd

d = {'one' : pd.Series([1, 2, 3], index = ['a', 'b', 'c']),
    'two' : pd.Series([1, 2, 3, 4], index = ['a', 'b', 'c', 'd'])}

df12 = pd.DataFrame(d)
```

```python
df12['three'] = pd.Series([10, 20, 30], index = ['a', 'b', 'c'])

df12
```

```python
df12['four'] = df12['one'] + df12['three']
df12
```

<h3><b>Column Deletion</b></h3>

Columns can be Deleted or popped From Existing DataFrame.


```python
import pandas as pd

d = {'one' : pd.Series([1, 2, 3], index = ['a', 'b', 'c']),
    'two' : pd.Series([1, 2, 3, 4], index = ['a', 'b', 'c', 'd']),
    'three' : pd.Series([10, 20, 30], index = ['a', 'b', 'c'])}

df13 = pd.DataFrame(d)
df13

# using del function
del df13['one']
df13

# use pop function
df13.pop('two')
df13
```

<h3 align='center'><b>Python Pandas - Indexing and Selecting Data</b></h3>

The Python and NumPy indexing operators "[ ]" and attribute operator "." provide quick and easy

optimization limits. For production code, we recommend that you take advantage of the

Pandas now supports three types of Multi-axes indexing; the three types are mentioned in the
following table 

<h4><b>.loc()</b></h4>

Pandas provide various methods to have purely <b>label based indexing.</b> When slicing, the end bound is also included. Integers are valid labels, but they refer to the label and not the position.

<b>.loc()</b> has multiple access methods like -
- A single scalar label
- A list of labels
- A slice object
- A Boolean array

<b>loc</b> takes two single/list/range operator seperated by ','. The first one indicates the row and the second one indicates columns.

```python
import pandas as pd

d = {'Name' : pd.Series(['Tom', 'James', 'Ricky', 'Vin', 'Steve', 'Smith', 'Jack', 'Lee', 'David', 'Gasper',
                        'Betina', 'Adres']),
    'Age' : pd.Series([25, 35, 45, 55, 23, 22, 20, 18, 24, 19, 20, 22]),
    'Rating' : pd.Series([0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0, 5.5, 6.0])}

df15 = pd.DataFrame(d)
df15
```

Select All Rows For a Specific column

```python
df15.loc[:, 'Name']
```

Select All rows for multiple columns from DataFrame

```python
df14.loc[:, ['Name', 'Age']]
```

Select Few Rows For Multiple Columns From DataFrame

```python
df14.loc[[0, 2, 4, 6], ['Name', 'Rating']]
```

Select Range of rows for all columns

```python
df15.loc[0:6]
```

For getting values with a boolean array

```python
df15.loc[0, ['Age', 'Rating']] > 0
```

<h3 align='CENTER'>HOW TO ADD ROWS TO DATAFRAME</h3>

```python
import pandas as pd
#List of Tuples
students = [('Jack', 34, 'Sydeny', 'Australia'),
           ('Riti', 30, 'Delhi', 'India'),
           ('Vikas', 31, 'Mumbai', 'India'),
           ('Nita', 24, 'Banglore', 'India'),
           ('John', 16, 'New York', 'Us'),
           ('Mike', 17, 'Las Vegas', 'US')]

students
```

```python
# Create a DataFrame From List of Tuples

df16 = pd.DataFrame(students, columns = ['Name', 'Age', 'City', 'Country'])

df16
```

<h4>Using DataFrame.append() and Dictionary</h4>
